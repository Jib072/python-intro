# set color value 1.0 to white

* https://matplotlib.org/stable/users/explain/colors/colorbar_only.html#sphx-glr-users-explain-colors-colorbar-only-py
* https://matplotlib.org/stable/api/_as_gen/matplotlib.colors.TwoSlopeNorm.html#matplotlib.colors.TwoSlopeNorm

```python
import matplotlib
```

```python
norm_colors=matplotlib.colors.TwoSlopeNorm(vmin=0, vcenter=1, vmax=3)

plt.matshow(data, origin='lower', cmap='bwr', norm = norm_colors)
plt.xticks(np.arange(len(labels_genes)), labels=labels_genes)
plt.yticks(np.arange(len(labels_conditions)), labels=labels_conditions)

cbar=plt.colorbar()
cbar.ax.set_ylabel('noramalised\ngene exprerssion')
```

# put xlabels to bottom (45 degree rotated)

```python
norm_colors=matplotlib.colors.TwoSlopeNorm(vmin=0, vcenter=1, vmax=3)
fig=plt.figure()
axheat = plt.gca()

heatmap=axheat.matshow(data, origin='lower', cmap='bwr', norm = norm_colors)
axheat.set_xticks(np.arange(len(labels_genes)), labels=labels_genes, rotation=45, ha='right', rotation_mode="anchor")
axheat.set_yticks(np.arange(len(labels_conditions)), labels=labels_conditions)
axheat.xaxis.set_ticks_position("bottom")

cbar=plt.colorbar(heatmap)
cbar.ax.set_ylabel('noramalised\ngene exprerssion')
```
