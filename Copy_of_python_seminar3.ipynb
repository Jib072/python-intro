* save as "AB_line.pdf"

# Errorbars
* add errorbars of 0.05 to every measurement of activity
* make markers smaller (8), include capsize of 4
* save new plot as AB_uncertainties.png with dpi 300

```python
#solution
fig = plt.figure(figsize=(4, 4))
plt.errorbar(x, ya, yerr=0.05, ms=8, color='red', marker='s', label='gene A', capsize=4)

plt.errorbar(x, yb, yerr=0.05, ms=8, color='green', marker='x', label='gene B', capsize=4)

plt.xlabel('concentration [mol/l]', fontsize=12, weight='bold', color='blue')
plt.ylabel('activity gene A', fontsize=12, weight='bold',color='blue')
plt.xlim([0,0.11])
plt.ylim([0,1.1])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(loc='lower right', fontsize =14)
plt.axhline(y=1.0, linestyle=':', color='gray')
plt.text(0.01, 1.02 , 'maximum', fontsize=12, color='gray')

plt.tight_layout()
plt.savefig('activities_AB.pdf')
```

# combining two plots
* create two empty subplots above each other figsze 4x6 inch, shared y axis
* documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
fig, axes = plt.subplots(2,1, figsize=(4, 4), sharex=True)
fig.subplots_adjust(hspace=0)
plt.show()
```

* scatter plot with gene A on top, an gene B bottom with style as previously
* save pdf named "AB_subplots.pdf"

```python
#solution:
fig, axes = plt.subplots(2,1, figsize=(4, 5), sharex=True)
axes[0].errorbar(x, ya, yerr=0.05, ms=8, color='red', marker='s', label='gene A', capsize=4)
axes[1].errorbar(x, yb, yerr=0.05, ms=8, color='green', marker='x', label='gene B', capsize=4)
axes[1].set_xlim([0,0.11])
axes[0].set_ylim([0,1.1])
axes[1].set_ylim([0,1.1])
axes[1].set_xlabel('concentration [mol/l]', fontsize=12, weight='bold')
axes[0].set_ylabel('activity gene A', fontsize=12, weight='bold')
axes[1].set_ylabel('activity gene B', fontsize=12, weight='bold')
plt.tight_layout()
fig.subplots_adjust(hspace=0) #has to be last so tight layout does not break subplot spacing
plt.savefig('AB_subplots.pdf')
```

# Venn diagram

* find python package which creates venn diagram, install it
* plot Venn digram for:
  - Human only: 4
  - Mouse only: 8
  - Human and Mouse: 12
  - Rat only: 1
  - Human and Mouse: 6
  - Mouse and Rat: 10
  - all: 9
* add title Human-Mouse-Rat gene comparison
* save as Venn.pdf

```python
#installation:
! pip install matplotlib_venn
```

```python
#solution:
from matplotlib_venn import venn3
```

```python
# Make the diagram
venn3(subsets = (4, 8, 12, 1, 6, 10, 9), set_labels = ('Human', 'Mouse', 'Rat'))
plt.title("Human-Mouse-Rat gene comparison")
plt.savefig('Venn.pdf')
```
